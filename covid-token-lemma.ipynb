```python
!pip install https://github.com/oroszgy/spacy-hungarian-models/releases/download/hu_core_ud_lg-0.3.1/hu_core_ud_lg-0.3.1-py3-none-any.whl  

import hu_core_ud_lg

nlp = hu_core_ud_lg.load()
```

```python
allowed = ['ADV', 'NOUN', 'ADJ', 'NUM', 'PROPN']

kommentek = []
for txt in df["Content of posts"]:
    doc = nlp(txt)
    proc = []
    for token in doc:
        if token.is_alpha and token.pos_ in allowed and not token.is_stop:
            proc.append(str(token.lemma_).lower())
    txt = ' '.join(proc)
    kommentek.append(txt)
```

```python
df["kommentek"] = kommentek
```

```python
df.to_csv("tokenized_comments_noverb.csv", index=False)
```
