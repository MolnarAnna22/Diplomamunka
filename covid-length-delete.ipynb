Már csak 2,4 millió komment maradt

```python
table = df_ossz.replace(r'\n', ' ', regex=True)
```

```python
table["Content of posts"][0]
```

```python
index = []
for i in range(len(table["Content of posts"])):
    if pd.isnull(table["Content of posts"][i]):
        index.append(i)
```

```python
table = table.drop(index)
table = table.reset_index(drop = True)
table.shape
```

```python
import re

text_nopunct = []
for i in table["Content of posts"]:
    text_nopunct.append(re.sub(r"[^a-záéíóöőúüűA-ZÁÉÍÓÖŐÚÜŰ0-9]+", ' ', i))
```

```python
len(text_nopunct)
```

```python
text_nopunct[0]
```

```python
table["text_nopunct"] = text_nopunct
```

```python
table.head()
```

```python
table.to_csv(r'table_nopunct.csv', index = False)
```

```python
table["text_nopunct"].to_csv(r'series_nopunct.csv', index = False)
```

## Kommentek hosszúsága (hány szóból állnak)

```python
txt_length = []
for i in range(len(table["text_nopunct"])):
    txt_length.append(len(str(table["text_nopunct"][i]).split()))
```

```python
table['txt_length'] = txt_length
```

```python
table.head(10)
```

```python
table['hunspell_ratio'] = table['hunspell']/table['txt_length']
```

```python
table = table[table["Domain"] == "www.facebook.com"]
```

```python
table.shape
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

fig, ax = plt.subplots()
sns.distplot(table["hunspell_ratio"], ax=ax, hist=True, kde=False, 
             bins=np.arange(0, 1.1, 0.1), color = 'blue',
             hist_kws={'edgecolor':'black'})
ax.set_ylim(0, 450000)
ax.set_yticks(range(0,450000,20000))
ax.set_xlim(0, 1.0, 0.1)
ax.set_xticks(np.arange(0, 1.1, 0.1))
plt.show()
```
